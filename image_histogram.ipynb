```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
orig_img = cv2.imread('testing_photos/uCAM_image_23-22-27_2018-11-30.jpg', cv2.IMREAD_UNCHANGED)
orig_img = cv2.resize(orig_img, (160, 120))
plt.imshow(orig_img)
plt.show()
```

```python
gray = cv2.cvtColor(orig_img, cv2.COLOR_RGB2GRAY)
# img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(gray)
plt.show()
gray
```

```python
plt.hist(gray.ravel(), 8, [0, 256])
plt.show()
```

```python
gray.shape
```

```python
pixels = [0]*8
print(pixels)

for y in range(gray.shape[0]):
    for x in range(gray.shape[1]):
        for i in range(8):
            if (i * 32) <= gray[y][x] <= (i * 32)+31:
                pixels[i] += 1
print(pixels)

# plt.hist()
plt.plot(range(8), pixels)
plt.show()
```

```python
nbins=8
pixels = [0]*nbins
print(pixels)

for pixel in gray.flatten():
    for i in range(nbins):
        if (i * (256 / nbins)) <= pixel <= (i * (256 / nbins))+((256 / nbins) - 1):
            pixels[i] += 1
print(pixels)


# plt.hist(pixels)
plt.plot(range(nbins), pixels)
plt.show()
```
